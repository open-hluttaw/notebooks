```python
#Getting Info on All Representatives in Amyotha

import requests #for REST requests to API
import pandas #for working with data
```

```python
#Get JSON of all posts and memberships in Amyotha:
amyotha_req = requests.get('http://api.openhluttaw.org/en/organizations/897739b2831e41109713ac9d8a96c845')

#Get details of all members of parliament
MPs = amyotha_req.json()['result']['memberships']

    
#List all MP Posts
constituencies = amyotha_req.json()['result']['posts']

for i in constituencies:
    print i['label']
```

```python
#MPs memberships also hold details of person and post of MP of a constituent
for mp in MPs:
    print mp['person']['name'] +" MP for " + mp['post']['label']
```

```python
#Persons and Posts also have other details such as birth_dates, gender etc.
#see popolo spec 
for mp in MPs:
    print mp['person']['name'] + " , " + mp['person']['gender'] + " , " +  mp['person']['birth_date']
```

```python
#To get contact details, we need to make additional requests for the person based on their id
for mp in MPs:
    person = requests.get('http://api.openhluttaw.org/en/persons/' + mp['person']['id']).json()['result']
    if person['contact_details']: #not everyone may contact details
        for contact in person['contact_details']: #can be multiple contact types
            if contact['type'] == 'email':
                print contact['value']
```

```python
#Example of CSV contact list of Women MPs

women = [ mp for mp in MPs if mp['person']['gender'] == 'female' ]

woman_mps = []

for mp in women:
    
    person = requests.get('http://api.openhluttaw.org/en/persons/' + mp['person']['id']).json()['result']
    
    woman_mp = {} 
    woman_mp ['name'] = mp['person']['name']    
    woman_mp ['constituency'] = mp['post']['label']
    
    if person['contact_details']: #not everyone may contact details
        for contact in person['contact_details']: #can be multiple contact types
            if contact['type'] == 'cell':
                woman_mp['phone'] = contact['value']
            elif contact['type'] == 'email':
                woman_mp['email'] = contact['value']
    woman_mps.append(woman_mp)

```

```python
women_df = pandas.DataFrame(woman_mps,columns=['name','constituency','phone','email'])

women_df
```

```python
women_df.to_csv() # eg. to save to file: print >>open('women_mp_contacts','a') , women_df.to_csv()
```
